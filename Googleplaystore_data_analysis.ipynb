# Question 3 Find out popular app which is updated before 2016. Popularity is based on no of installs

```python
df3 = pd.read_csv("googleplaystore.csv")
```

```python
index = df3[df3["Category"] == '1.9'].index
df3.drop(index, inplace=True)
```

```python
df3
```

```python
df3['Installs'] = df3['Installs'].str.replace('+', '')
df3['Installs'] = df3['Installs'].str.replace(',', '')
df3["Installs"] = df3["Installs"].replace("Free",np.nan)
```

```python
df3.rename(columns = {"Last Updated": "Last_Updated"},  
          inplace = True)
```

```python
df3
```

```python
df3['Last_Updated'] = df3['Last_Updated'].str.replace(',', '')
df3['Last_Updated'] = df3['Last_Updated'].str.replace('1.0.19', 'January 1, 2019')
```

```python
df3['Last_Updated'] = df3['Last_Updated'].str.replace('January 1, 2019', 'January 1 2019')
```

```python
df3 = df3[["App","Installs", "Last_Updated"]]
```

```python
df3
```

```python
df3["Last_Updated"] = [datetime.strptime(i, '%B %d %Y') for i in df3["Last_Updated"]]
```

```python
df3
```

```python
df3['Last_Updated'] = pd.to_datetime(df3['Last_Updated'],format = '%Y-%m-%d ').dt.year
```

```python
df3
```

```python
df3 = df3[df3.Last_Updated < 2016]
```

```python
df3
```

```python
df3 = df3.sort_values(["Installs"], ascending=[False])
```

```python
df3.head(10)
```

# ANSWER 3: the contents of the above table shows the apps having the most popularity based on installs which were installed before 2016
